さて，いまから， $y=\sin(x)$, $x\in[0,5]$のグラフを描いてみます．<br>
基本的な考え方は，$[0,5]$を細かく分割し，それぞれの点で$\sin$の値を計算し，それらを結んでいくというものです．<br>
そこで，$[0,5]$を細かく分割した配列（ここでは$[0.0,0.01,0.02,\dots,4.49,5.0]$という配列）を定義しましょう．

```julia
x = 0:0.01:5
# length(x)
```

ハッシュタグを外して，length(x)を表示させると，501 になるはずです．

次に，それぞれの値に対して$\sin$の値を計算し，それをyという名前の配列に格納しましょう．<br>
このとき，sin(x)とするとエラーが発生します．通常，sin()と書くと，括弧の中はスカラーである必要があるのに，いまは配列だからです．こういうときは，sinの後ろにドットを書いて対処します．
https://docs.julialang.org/en/v1/manual/mathematical-operations/#man-dot-operators

```julia
y = sin.(x)
```

さて，関数をプロットする準備が整いました．最小限，<br> 
　plot(x,y) <br>
でグラフは描けます．yより後ろはオプションです．

```julia
plot(x, y, xlabel="x", ylabel="y", label="y=sin(x)")
```

同じグラフに別の関数をプロットしてみましょう．こういうときは，plotの直後にビックリマーク!を書き入れます．

```julia
z = cos.(x)*2 - 1/10*x.^2
plot!(x, z, label="y=cos^2(x)-1/10*x^2")
```
